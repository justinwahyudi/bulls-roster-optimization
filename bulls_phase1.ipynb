```python
import pandas as pd
import matplotlib.pyplot as plt

# Load from saved CSVs — no API call needed
df_all_players = pd.read_csv('nba_all_players_2526.csv')
df_bulls = pd.read_csv('bulls_roster_stats_2526.csv')

print("Players in league dataset:", len(df_all_players))
print("Bulls players:", len(df_bulls))
```

```python
# Sort by blocks and show top players
df_bulls_sorted = df_bulls.sort_values('BLK', ascending=False)
print(df_bulls_sorted[['PLAYER_NAME', 'BLK']])
```

```python
# Blocks per game chart
plt.figure(figsize=(10, 6))
plt.barh(df_bulls_sorted['PLAYER_NAME'], df_bulls_sorted['BLK'])
plt.xlabel('Blocks Per Game')
plt.ylabel('Player Name')
plt.title('Bulls Player Stats - Blocks Per Game')
plt.tight_layout()
plt.savefig('bulls_blocks_chart.png')
plt.show()
```
